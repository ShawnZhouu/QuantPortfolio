```python
import pandas as pd
import numpy as np
from BTC_Alpha_func import *
from tqdm import tqdm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def single_factor_backtest(fac, ret, thresh=10, freq=5):
    all_dates = fac.index.get_level_values(0).unique()
    all_returns = [0]
    long_ = []
    short_ = []
    long_stock_  = []
    short_stock_ = []
    for date in tqdm(all_dates[::freq]):
        this_date_factor = fac.loc[date]
        this_date_retrun = ret.loc[date]
        long_thresh = np.percentile(this_date_factor, 100-thresh)
        short_thresh = np.percentile(this_date_factor, thresh)
        long_stock = this_date_retrun[this_date_factor >= long_thresh].dropna().index
        short_stock = this_date_retrun[this_date_factor <= short_thresh].dropna().index
        long_return = np.mean(this_date_retrun[this_date_factor >= long_thresh]).values[0]
        short_return = np.mean(this_date_retrun[this_date_factor <= short_thresh]).values[0]
        if short_return < -1:
            short_return = -1
        if long_return < -1:
            long_return = -1
        all_returns.append(long_return - short_return)
        long_.append(long_return)
        short_.append(short_return)
        long_stock_.append(long_stock)
        short_stock_.append(short_stock)
    all_returns = np.array(all_returns)
    return np.nan_to_num(all_returns), np.nan_to_num(long_), np.nan_to_num(short_), all_dates[::freq], \
            np.nan_to_num(long_stock_), np.nan_to_num(short_stock_)

def performance(capitals, all_dates, long_stock, short_stock, factor_name):
    plt.figure(figsize=(20,10))
    capital_list = np.cumprod(capitals+1)
    plt.plot(all_dates, capital_list[1:])
    plt.title(factor_name)
    ret = ret_ratio(capital_list)
    sr = sharpe_ratio(capital_list)
    md = max_dd(capital_list)
    to = turnover(long_stock, short_stock)
    print('Return:', ret)
    print('Sharpe Ratio:', sr)
    print('Maximum Drawdown:', md)
    print('Turnover:', to)
    print('Score:', sr*np.sqrt(ret/to))
    
def ret_ratio(capital_list):
    r=(1+(capital_list[-1] - capital_list[0]) / capital_list[0])**(250/len(capital_list))-1
    return r

def sharpe_ratio(capital_list):
    return ret_ratio(capital_list) / (np.std(capital_list)*np.sqrt(250))
    
def max_dd(capital_list):
    i = np.argmax((np.maximum.accumulate(capital_list) - capital_list) / np.maximum.accumulate(capital_list)) # 结束位置 
    if i == 0: return 0
    j = np.argmax(capital_list[:i]) # 开始位置
    return (capital_list[j] - capital_list[i]) / (capital_list[j])

def turnover(long, short):
    turn_over = []
    for i in range(1, len(long)):
        if len(long[i-1]) == 0:
            long_to = 1
        else:
            long_to = len(set(long[i-1]) - set(long[i])) / len(long[i-1])
        if len(short[i-1]) == 0:
            short_to = 1
        else:
            short_to = len(set(short[i-1]) - set(short[i])) / len(short[i-1])
        turn_over.append(long_to + short_to)
    return np.mean(turn_over) / 2
```

```python
final_data = pd.read_parquet('v2.parquet')
ret = final_data[['adj_close']].groupby('stock').apply(lambda x:(DELAY(x,-5)/x-1).fillna(0))
```

```python
factors = pd.concat([-momentum, vol_adv20, -vol_open_close_factor, abs_ret_sum_factor, neg_close_factor, 
                    highday_lowday_factor, -short_debt_equity_factor, -sales_asset_factor, -EBITDA_asset_factor,
                    total_asset_factor, -return_on_equity_factor, sales_inventory_factor], axis=1)
```

# PRICE & VOLUME

```python
momentum = final_data[['adj_close']].groupby('stock').apply(lambda x:(x/DELAY(x,5)-1).fillna(0))
capital, long, short, all_dates, long_stock, short_stock = single_factor_backtest(-momentum, ret, thresh=20, freq=5)
performance(capital, all_dates, long_stock[1:], short_stock[1:], 'Negative Momentum')
```

```python
def vol_adv20(data):
    return data['adj_volume'].groupby('stock').apply(lambda x:x/SUM(x, 20))

vol_adv20 = vol_adv20(final_data)
capital, long, short, all_dates, long_stock, short_stock  = single_factor_backtest(vol_adv20, ret, 
                                                                                   thresh=20, freq=5)
performance(capital, all_dates, long_stock[1:], short_stock[1:], 'VOL_ADV20')
```

```python
def vol_open_close(data):
    fac = (data['adj_volume'] / (data['adj_open'] - data['adj_close']))
    return fac.groupby('stock').apply(lambda x:EWM(x,20))

vol_open_close_factor = vol_open_close(final_data)
capital, long, short, all_dates, long_stock, short_stock  = single_factor_backtest(-vol_open_close_factor, ret, 
                                                                                   thresh=20, freq=5)
performance(capital, all_dates, long_stock[1:], short_stock[1:], 'VOL_OPEN_CLOSE_20')
```

```python
def abs_ret_sum(data):
    # 20-abs-return-sum
    ret = data.groupby('stock')['adj_close'].apply(lambda x:x/DELAY(x, 1)-1)
    fac = ret.groupby('stock').apply(lambda x: EWM(ABS(x),20))
    return fac  

abs_ret_sum_factor = abs_ret_sum(final_data)
capital, long, short, all_dates, long_stock, short_stock  = single_factor_backtest(abs_ret_sum_factor, ret, 
                                                                                   thresh=20, freq=5)
performance(capital, all_dates, long_stock[1:], short_stock[1:], 'ABS_RET_EWM_SUM_20')
```

```python
def neg_close(data):
    return -data['adj_close']

neg_close_factor = neg_close(final_data)
capital, long, short, all_dates, long_stock, short_stock  = single_factor_backtest(neg_close_factor, ret, 
                                                                                   thresh=20, freq=5)
performance(capital, all_dates, long_stock[1:], short_stock[1:], 'NEG_CLOSE')
```

```python
def highday_lowday(data):
    # 20-abs-return-sum
    fac = data['adj_close'].groupby('stock').apply(lambda x:(HIGHDAY(x,20)-LOWDAY(x,20))/20).fillna(0)
    return fac  

highday_lowday_factor = highday_lowday(final_data)
capital, long, short, all_dates, long_stock, short_stock  = single_factor_backtest(highday_lowday_factor, ret, 
                                                                                   thresh=20, freq=5)
performance(capital, all_dates, long_stock[1:], short_stock[1:], 'HIGHDAY_LOWDAY_20')
```

# FUNDAMENTAL

```python
def short_debt_equity(data):
    #3051/3501
    fac = (data['ITEM3051'] / (data['ITEM3501']+0.0001))
    return fac

short_debt_equity_factor = short_debt_equity(final_data)
capital, long, short, all_dates, long_stock, short_stock  = single_factor_backtest(-short_debt_equity_factor, ret, 
                                                                                   thresh=20, freq=5)
performance(capital, all_dates, long_stock[1:], short_stock[1:], 'SHORT_DEBT_EQUITY')
```

```python
def sales_asset(data):
    #18198/2201
    #EBITDA/ASSET
    fac = TSRANK(data['ITEM18198'] / data['ITEM2201'], 10)
    return fac

sales_asset_factor = sales_asset(final_data)
capital, long, short, all_dates, long_stock, short_stock  = single_factor_backtest(-sales_asset_factor, ret, 
                                                                                   thresh=20, freq=5)
performance(capital, all_dates, long_stock[1:], short_stock[1:], 'SALES_ASSETS')
```
